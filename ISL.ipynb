```python
from islpy import Set
from islplot.plotter3d import plot_set_3d
from IPython.display import HTML
```

```python
import islpy as isl

space = isl.Space.create_from_names(isl.DEFAULT_CONTEXT, set=["L", "U", "I"])

bset = (isl.BasicSet.universe(space)
        .add_constraint(isl.Constraint.ineq_from_names(space, {1: -1, "L": 1}))
        .add_constraint(isl.Constraint.ineq_from_names(space, {1: 10, "L": -1}))
        .add_constraint(isl.Constraint.ineq_from_names(space, {"U": 1, "L": -1}))
        .add_constraint(isl.Constraint.ineq_from_names(space, {1: 10, "U": -1}))
        .add_constraint(isl.Constraint.ineq_from_names(space, {"I": 1, "L": -1}))
        .add_constraint(isl.Constraint.ineq_from_names(space, {"I": -1, "U": 1})))
print("set 1 %s:" % bset)
```

```python
s = Set("{ [L, U, I] : 0 < L <= 10 and L <= U <= 10 and L <= I <= U }")
d = plot_set_3d(s, show_points=True, show_vertices=True,show_axes=True)
HTML(d)
```
